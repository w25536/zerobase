```python
# 3-1
def find_code(ingredient):
    code = ingredients_df[ingredients_df['표준 영문명'].str.lower() == ingredient.lower()]['성분코드'].values # 표준 영문명에 대한 성분코드를 찾아서 반환
    if len(code) > 0: # code가 존재하면
        return code[0] # code의 첫번째 값을 반환
    
    code = ingredients_df[ingredients_df['구영문명'].str.lower() == ingredient.lower()]['성분코드'].values # 구영문명에 대한 성분코드를 찾아서 반환
    if len(code) > 0:
        return code[0]    
    
    return None

df_target['Code List'] = df_target['Ingredients List'].apply(lambda x: [find_code(ingredient) for ingredient in x]) 

df_target[['Ingredients List', 'Code List']].head()
check_03_01(df_target)
df_target
```
