위도, 경도 정보를 이용하여 미왕빌딩에서 셀프주유가 가능하고 24시간이면서 편의점이 있는 가장 가까운 주유소 10개를 휘 발유 가격이 가장 저렴한 순으로 정렬하여 조회하세요.

- 주유소 아이디, 주유소 브랜드명, 주유소 매장명, 주소, 휘발유 가격, 부가정보 (셀프, 24시간, 편의점 여부), 미왕빌딩으로 부터의 거리 (km)


제출 7.
- 쿼리, 결과

```python
sql_query_q7 = """
SELECT s.id, b.name AS brand_name, s.name AS station_name, s.address, s.gasoline AS gasoline_price,
       CASE WHEN s.self = 1 THEN 'Yes' ELSE 'No' END AS self_service,
       CASE WHEN s.24_hours = 1 THEN 'Yes' ELSE 'No' END AS 24_hours,
       CASE WHEN s.convenience_store = 1 THEN 'Yes' ELSE 'No' END AS convenience_store,
       (6371*acos(cos(radians(37.4955366))*cos(radians(s.lat))*cos(radians(s.lng)
           -radians(127.0293521))+sin(radians(37.4955366))*sin(radians(s.lat)))) AS distance 
FROM GAS_BRAND b
JOIN GAS_STATION s ON s.brand = b.id
WHERE s.self = 1 AND s.24_hours = 1 AND s.convenience_store = 1
HAVING distance <= 10
ORDER BY s.gasoline ASC, distance ASC
LIMIT 10;
"""

cur = conn.cursor()
cur.execute(sql_query_q7)
sql_result_q7 = cur.fetchall()
cur.close()

for row in sql_result_q7:
    print(row)
```

구별로 주유소 브랜드 별 휘발유 평균가격을 조회하여 저렴한 순으로 출력하세요.

- 구 이름, 주유소 브랜드 이름, 휘발유 평균 가격

제출 8.
- 쿼리, 결과

```python
sql_query_q8 = """
SELECT s.gu, b.name, AVG(s.gasoline) AS avg_gasoline_price
FROM GAS_STATION s
JOIN GAS_BRAND b ON s.brand = b.id
GROUP BY s.gu, b.name
ORDER BY avg_gasoline_price ASC; 
"""

cur = conn.cursor()
cur.execute(sql_query_q8)
sql_result_q8 = cur.fetchall()
cur.close()

for row in sql_result_q8:
    print(row)
```
